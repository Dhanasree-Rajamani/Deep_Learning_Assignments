Performing classification to classify images as monkey or not

Training dataset consists of images of Monkeys from Kaggle. Test to perform checks done with tiger, monkey and elephant images and got accurate predictions.

Using Fast AI - vision learner to perform classification 

Using google drive here as github could not hold images of this size

```python
from fastai.vision.all import *
```

Classification

If image has monkey : return True, if image has something else : return False

Using FastAI library - vision_learner to perform classification

```python
!unzip '/content/gdrive/MyDrive/training.zip' -d '.'
```

```python
def is_monkey(x): return x[0].isupper()

#path, list of images, labels as input
#10% validation 90% for training 
dls = ImageDataLoaders.from_name_func(
    '.', get_image_files('.'), valid_pct=0.1, seed=42,
    label_func=is_monkey, item_tfms=Resize(224))

learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
```

```python
from PIL import Image
import requests, io, cv2
```

Predicting whether given image is monkey or not monkey

```python
#hide
response = requests.get('https://cdn.britannica.com/02/152302-050-1A984FCB/African-savanna-elephant.jpg')
bytes_im = io.BytesIO(response.content)
elephant = cv2.cvtColor(np.array(Image.open(bytes_im)), cv2.COLOR_RGB2BGR)
```

```python
img = PILImage.create(elephant)
is_monkey,_,probs = learn.predict(img)
print(f"Is this a Monkey?: {is_monkey}.")
print(f"Probability it's a monkey: {probs[1].item():.6f}")
```
